# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#1. Get the total rides for each city type
total_rides_by_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_city_type
```

```python
# 2. Get the total drivers for each city type
sum_drivers_by_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
sum_drivers_by_city_type
```

```python
#  3. Get the total amount of fares for each city type
sum_fares_by_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
sum_fares_by_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_by_city_type = (sum_fares_by_city_type / total_rides_by_city_type)
avg_fare_by_city_type
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_driver = (sum_fares_by_city_type / sum_drivers_by_city_type)
avg_fare_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary = {"Total Rides": (total_rides_by_city_type), "Total Drivers": (sum_drivers_by_city_type), "Total Fares": (sum_fares_by_city_type), "Average Fare Per Ride": (avg_fare_by_city_type), "Average Fare Per Driver": (avg_fare_driver)}
pyber_summary_df = pd.DataFrame(pyber_summary)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df.style.format({'Total Fares':'${0:,.2f}',
                              'Average Fare per Ride':'${:.2f}',
                              'Average Fare per Driver':'${:.2f}',
                              'Total Drivers':'{:,}',
                              'Total Rides':'{:,}'})
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 9. Using groupby() on the pyber_data_df to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fare_by_date_type_df = pyber_data_df.groupby(['type','date']).sum()["fare"]
fare_by_date_type_df = pd.DataFrame(fare_by_date_type_df)       
fare_by_date_type_df
```

```python
# 10. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
reset_index = fare_by_date_type_df.reset_index()
fare_by_date_type_df
```

```python
# 11. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fare_by_date_type_df = reset_index.pivot(index="date", columns="type")["fare"]
fare_by_date_type_df
```

```python
# 12. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2018-01-01':'2018-04-29'.
date_given = fare_by_date_type_df.loc["2019-01-01" :"2019-04-29"]
date_given.head()
```

```python
# 13. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
fare_by_date_type_df.index = pd.to_datetime(fare_by_date_type_df.index)
fare_by_date_type_df
```

```python
# 14. Check that the datatype for the index is datetime using df.info()
date_given.info()
```

```python
# 15. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fare_by_date_type_df_re = fare_by_date_type_df.resample("W").sum()
fare_by_date_type_df_re
```
